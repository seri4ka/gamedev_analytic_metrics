```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_cpi(ad_spend, installs):
    """Вычисление метрики CPI.
    
    Args:
        ad_spend (float): Рекламные расходы.
        installs (int): Количество установок.
    
    Returns:
        float: CPI.
    """
    if installs == 0:
        raise ValueError("Количество установок не может быть равно нулю.")
    return ad_spend / installs
```

```python
data_path = "../data/cpi_data.csv"
data = pd.read_csv(data_path)
data
```

```python
# Расчет CPI для каждой кампании
data["CPI"] = data.apply(lambda row: calculate_cpi(row["ad_spend"], row["installs"]), axis=1)

data.index = data['ad_campaign']
data.iloc[:, 1:].plot(subplots=True)
plt.show()
```
